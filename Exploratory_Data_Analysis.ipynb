```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
xtrain=pd.read_csv("xtrain.csv")
xtrain
```

```python
ytrain = pd.read_csv("ytrain.csv")
ytrain
```

```python
xtrain["TARGET"] = ytrain["TARGET"]
xtrain
```

```python
df = pd.DataFrame(columns=["ID","Timestamp","Value","Target"])
df
```

```python
for i in xtrain.index[:20]: 
    k = 0
    for j in xtrain.columns[2:] :
        print(i)
        print(j)
        print(xtrain[j][i])
        df = df.append({"ID" : xtrain["ID"][i], "Timestamp" : k, "Value" : xtrain[j][i], "Target" : xtrain["TARGET"][i]}, ignore_index = True)
        k+=1
```

```python
df
```

```python
fig = px.line(df, x="Timestamp", y="Value", color='Target', title = "Séquences de pics neuronaux")
fig.show()
```
