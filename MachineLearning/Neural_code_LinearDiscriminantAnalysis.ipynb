Dans ce notebook, on applique une analyse discrimante linéaire sur le jeu de données des caractéristiques (features) extraites des séries temporelles et le jeu de données des caractéristiques (features) extraites des séries temporelles différenciées. On a choisi d'essayer cet algorithme de machine learning, car on a un jeu de données en très grande dimension, ce qui permet d'appliquer une réduction de dimension.

```python
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import cohen_kappa_score, make_scorer
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

**Avec le dataset xtrain_features**
dataset des caractéristiques extraites de la série à l'aide du package tsfresh

```python
xtrain_features = pd.read_csv("xtrain_features.csv")
xtrain_features
```

```python
ytrain = pd.read_csv("ytrain.csv")
ytrain
```

```python
# création du modèle
model=LinearDiscriminantAnalysis(n_components = 1) # n_components : choix de 1 axe discriminant, car on a 2 classes
```

On fait une validation croisée avec l'échantillon d'apprentissage pour estimer le risque.

```python
LDA=cross_val_score(model, xtrain_features.iloc[:,1:], ytrain['TARGET'], cv=10, scoring = make_scorer(cohen_kappa_score))
LDA
```

```python
LDA.mean()
```

```python
Fact=model.fit_transform(xtrain_features.iloc[:,1:], ytrain['TARGET']) # calcule la valeur des individus sur les axes discriminants 
pd.DataFrame(Fact,columns=['X1']).head(5) # exemple des 5 premiers individus 
```

```python
# axes discriminants
dfFact = pd.DataFrame(Fact,columns=['F1'])
dfFact['classe'] = ytrain['TARGET']
dfFact['y'] = 0
dfFact
```

```python
#représentation graphique
sns.scatterplot(x="F1",y="y",hue="classe",data=dfFact,palette="deep")
```

Prédiction avec ce modèle sur le jeu de données xtest_features

```python
df = pd.read_csv("xtest_features.csv")
xtest_features= df.set_index("Unnamed: 0")
xtest_features
```

```python
y_test_pred = model.predict(xtest_features)
y_test_pred
```

```python
ytest = pd.DataFrame(np.transpose([y_test_pred]),columns=['TARGET'], index = xtest_features.index)
ytest
```

```python
ytest.to_csv("ytest_LDA.csv") # score de 0.12764062147842603 avec ce jeu de données
```

**Avec le dataset xtrain_features_diff**
dataset des caractéristiques extraites de la série différenciée à l'aide du package tsfresh

```python
xtrain_features_diff=pd.read_csv("xtrain_features_diff.csv")
xtrain_features_diff
```

```python
# création du modèle
model_diff=LinearDiscriminantAnalysis(n_components = 1) # n_components : choix de 2 axes discriminants
```

```python
LDA_diff=cross_val_score(model_diff, xtrain_features_diff.iloc[:,1:], ytrain['TARGET'], cv=10, scoring = make_scorer(cohen_kappa_score))
LDA_diff
```

```python
LDA_diff.mean()
```

```python
Fact_diff=model_diff.fit_transform(xtrain_features_diff.iloc[:,1:], ytrain['TARGET']) # calcule la valeur des individus sur les axes discriminants 
pd.DataFrame(Fact_diff,columns=['X1']).head(5) # exemple des 5 premiers individus 
```

```python
# axes discriminants
dfFact_diff = pd.DataFrame(Fact_diff,columns=['F1'])
dfFact_diff['classe'] = ytrain['TARGET']
dfFact_diff['y'] = 0 # pour le graph
dfFact_diff
```

```python
#représentation graphique
sns.scatterplot(x="F1",y="y",hue="classe",data=dfFact_diff,palette="deep")
```

**Prédiction sur le jeu de données xtest_features_diff pour faire une sousmission**
