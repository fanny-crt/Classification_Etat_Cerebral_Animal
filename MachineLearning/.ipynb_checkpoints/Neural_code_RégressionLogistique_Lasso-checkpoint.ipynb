Dans ce notebook, on applique une régression logistique penalisée (l1 : lasso) aux fichiers :  
- xtrain_features.csv : fichier d'extraction des caractéristiques/features des séries temporelles initiales (xtrain.csv) 
- xtrain_features_diff.csv : fichier d'extraction des caractéristiques/features des séries temporelles différenciées

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#jeu de couleurs pour faire joli
import matplotlib.cm as cm

from sklearn.preprocessing import StandardScaler

from sklearn.metrics import cohen_kappa_score, make_scorer 
from sklearn.model_selection import GridSearchCV

# Régression logistique
from sklearn.linear_model import LogisticRegression

# Lasso
from sklearn.linear_model import lasso_path
from sklearn.linear_model import Lasso

# sélection de variables
from sklearn.feature_selection import SelectFromModel
```

```python
ytrain = pd.read_csv("ytrain.csv")
ytrain
```

**Sur les features des séries temporelles**

```python
df = pd.read_csv("xtrain_features.csv")
xtrain_features = df.set_index("Unnamed: 0")
xtrain_features
```

**Sélection de variables avec la méthode lasso appliquée à la régression logistique**

On normalise les données

```python
sc = StandardScaler()
xtrain_features_std = sc.fit_transform(xtrain_features)
xtrain_features_std
```

Grille de recherche par validation croisée pour sélectionner le alpha pour le lasso

```python
params = {
    "C" : np.array([0.001,0.01,0.02,0.025,0.05,0.06,0.07,0.08,0.09,0.1,0.25,0.5,0.8,1.0]),
    "solver" : ["liblinear", "saga"]
} # C = les lambda testés du lasso

model = LogisticRegression(penalty="l1") # penalty = "l1" -> la méthode lasso appliquée à la régression logistique
grid = GridSearchCV(model, param_grid=params, cv=5, scoring = make_scorer(cohen_kappa_score))
```

```python
grid.fit(xtrain_features_std,ytrain['TARGET'])
```

```python
grid.best_params_
```

**Résultats :** {'C': 0.8, 'solver': 'liblinear'}

```python
grid.best_score_ # 0.15302995883286458
```

```python
RegLog = LogisticRegression(penalty = "l1", solver= "liblinear",C=0.8)
RegLog.fit(xtrain_features_std,ytrain['TARGET'])
```

```python
RegLog.coef_ # coefs de la régression logistique pénalisée avec alpha = 0.8
```

```python
nbNonZero = np.apply_along_axis(func1d=np.count_nonzero,arr=RegLog.coef_,axis=0)
nbNonZero.sum() # 175 variables "allumées" avec alphas = 0.8
```

```python
#nom des variables
nom_var = xtrain_features.columns

#affichage des coefficients pour alpha = 0.25
pd.DataFrame({'Variables':nom_var,'Coefficients': RegLog.coef_[0]})
```

Prédiction sur le jeu de données xtest_features

```python
df = pd.read_csv("xtest_features.csv")
xtest_features = df.set_index("Unnamed: 0")
xtest_features
```

On normalise les données

```python
xtest_std_features = sc.fit_transform(xtest_features)
xtest_std_features
```

Prédiction de ytest

```python
y_test_pred_features = RegLog.predict(xtest_std_features)
y_test_pred_features
```

```python
ytest_features = pd.DataFrame(np.transpose([y_test_pred_features]),columns=['TARGET'], index = xtest_features.index)
ytest_features
```

```python
ytest_features.to_csv("ytest_RegLogistique_features_Lasso.csv") # 0.13892146776506442
```

**Sur les features des différenciations des séries temporelles**

```python
df = pd.read_csv("xtrain_features_diff.csv")
xtrain_features_diff = df.set_index("Unnamed: 0")
xtrain_features_diff
```

**Sélection de variables avec la méthode lasso appliquée à la régression logistique**

On normalise les données

```python
sc = StandardScaler()
xtrain_features_diff_std = sc.fit_transform(xtrain_features_diff)
xtrain_features_diff_std
```

Grille de recherche par validation croisée pour sélectionner les paramètres

```python
params = {
    "C" : np.array([0.001,0.01,0.02,0.025,0.05,0.06,0.07,0.08,0.09,0.1,0.25,0.5,0.8,1.0]),
    "solver" : ["liblinear", "saga"]
} # C = les alphas du lasso

model = LogisticRegression(penalty="l1") # penalty = "l1" -> la méthode lasso appliquée à la régression logistique
grid_diff = GridSearchCV(model, param_grid=params, cv=5, scoring = make_scorer(cohen_kappa_score))
```

```python
grid_diff.fit(xtrain_features_diff_std,ytrain['TARGET'])
```

```python
grid_diff.best_params_
```

**Résultats :** {'C': 1.0, 'solver': 'liblinear'}

```python
grid_diff.best_score_ # 0.18231051126922088
```

```python
RegLog_diff = LogisticRegression(penalty = "l1", solver="liblinear" ,C=1)
RegLog_diff.fit(xtrain_features_diff_std,ytrain['TARGET'])
```

```python
RegLog_diff.coef_ # coefs de la régression logistique pénalisée avec alpha = 1
```

```python
nbNonZero = np.apply_along_axis(func1d=np.count_nonzero,arr=RegLog_diff.coef_,axis=0)
nbNonZero.sum() # 205 variables "allumées" avec alphas = 1
```

Prédiction sur le jeu de données xtest_features_diff

```python
df = pd.read_csv("xtest_features_diff.csv")
xtest_features_diff = df.set_index("Unnamed: 0")
xtest_features_diff
```

On normalise les données

```python
xtest_std_features_diff = sc.fit_transform(xtest_features_diff)
xtest_std_features_diff
```

Prédiction de ytest

```python
y_test_pred_features_diff = RegLog_diff.predict(xtest_std_features_diff)
y_test_pred_features_diff
```

```python
ytest_features_diff = pd.DataFrame(np.transpose([y_test_pred_features_diff]),columns=['TARGET'], index = xtest_features_diff.index)
ytest_features_diff
```
