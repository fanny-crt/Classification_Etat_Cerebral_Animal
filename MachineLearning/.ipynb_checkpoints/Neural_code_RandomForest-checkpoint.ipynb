Dans ce notebook, on effectue un random forest sur le jeu de données des caractéristiques (features) extraites des séries temporelles et le jeu de données des caractéristiques (features) extraites des séries temporelles différenciées.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import cohen_kappa_score, make_scorer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_csv("xtrain.csv")
xtrain=df.set_index("ID")
xtrain
```

```python
df=pd.read_csv("ytrain.csv")
ytrain=df.set_index("ID")
ytrain
```

**Random Forest sur l'échantillon de base**

On fait une validation croisée pour estimer le score de Cohen Kappa sur l'échantillon d'apprentissage

```python
model = RandomForestClassifier(n_estimators=100,min_samples_split=100)
classifier=cross_val_score(model, xtrain, ytrain['TARGET'], cv=10, scoring = make_scorer(cohen_kappa_score))
classifier
```

```python
classifier.mean() # score de cohen kappa très faible 
```

**Random forest sur les features**

```python
df = pd.read_csv("xtrain_features.csv")
xtrain_features = df.set_index('Unnamed: 0')
xtrain_features
```

Dans un premier temps, on effectue un random forest avec des paramètres par défaut (n_estimators : 100 et min_sample_split : 100). Afin de déterminer la qualité du modèle, on utilse la validation croisée. 

```python
model_features = RandomForestClassifier(n_estimators=100,min_samples_split=100)
classifier_features=cross_val_score(model_features, xtrain_features, ytrain['TARGET'], cv=10, scoring = make_scorer(cohen_kappa_score))
classifier_features
```

```python
classifier_features.mean()
```

```python
model_features.fit(xtrain_features, ytrain['TARGET'])
```

**Importance des colonnes en ordre décroissant**

On regarde l'importance des variables en ordre décroissant de l'algorithme random forest. 

```python
importance = model_features.feature_importances_
df_importance = pd.DataFrame(np.transpose([xtrain_features.columns,importance]), columns = ['colonnes',"importance"])
df_importance.sort_values(by=['importance'], ascending = False)
```

On garde les 30 premières colonnes. 

```python
importance_d=df_importance.sort_values(by=['importance'], ascending = False)
xtrain_features_imp=xtrain_features[importance_d.iloc[:30,0]]
xtrain_features_imp
```

**Grille de recherche par validation croisée**

On effectue une grille de recherche par validation croisée pour déterminer les paramètres de l'algorithme Random Forest. 

```python
params = {'n_estimators': [50,100,200,300],
         'criterion' : ['gini','entropy'],
         'min_samples_split':[10,25,50,75,100],
         'min_samples_leaf':[10,25,50,75,100]}
gs = GridSearchCV(RandomForestClassifier(), param_grid=params,
                  scoring = make_scorer(cohen_kappa_score),
                  n_jobs=-1, cv=5, return_train_score=True)
```

```python
gs.fit(xtrain_features_imp,ytrain['TARGET'])
```

```python
gs.best_params_
```

**Résultats** : {'criterion': 'entropy',
 'min_samples_leaf': 10,
 'min_samples_split': 10,
 'n_estimators': 50}

```python
gs.best_score_ # 0.130401824971694
```

**Prédiction sur les meilleurs paramètres**

```python
df = pd.read_csv("xtest_features.csv")
xtest_features = df.set_index('Unnamed: 0')
xtest_features_imp=xtest_features[importance_d.iloc[:30,0]]
xtest_features_imp
```

```python
y_test_pred = gs.predict(xtest_features_imp)
y_test_pred
```

```python
ytest = pd.DataFrame(np.transpose([y_test_pred]),columns=['TARGET'], index = xtest_features_imp.index)
ytest
```

```python
ytest.to_csv("ytest_RandomForest_features.csv") # score de 0.06867827531620763
```
