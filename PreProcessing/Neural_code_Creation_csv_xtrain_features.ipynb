Dans ce notebook, on crée :

**Création du dataframe formaté à la librairie tsfresh**

```python
t=[]
t.extend(np.arange(0,50))
time = t*16635
len(time)
```

**Différenciation**

On extrait également les features mais la différence avec l'extraction précédente est que l'on extrait les features du jeu de données des séries temporelles différenciées.

```python
xtrain_diff=xtrain.iloc[:,2:].T.diff()[1:].T
xtrain_diff
```

**Création du dataframe formaté à la librairie tsfresh et différenciée**

```python
t_diff=[]
t_diff.extend(np.arange(0,49))
time = t_diff*16635
len(time)
```

```python
ID_diff = []
for i in xtrain_diff.index :
    x=[i]*49
    ID_diff.extend(x)
len(ID_diff)
```

```python
Timeseries_diff=[]
for row in xtrain_diff.itertuples() : 
    Timeseries_diff.extend(row[1:])

len(Timeseries_diff)
```

```python
df_diff = pd.DataFrame(np.transpose([time,ID_diff,Timeseries_diff]),columns=['time','ID','timeseries'])
df_diff
```

```python
extracted_features = extract_features(df_diff, column_id="ID", column_sort="time")
```

```python
extracted_features
```

```python
df_diff=extracted_features.dropna(axis = 1)  # on supprime les colonnes avec des valeurs manquantes
df_diff
```

On filtre les features

```python
features_filtered_diff = select_features(df_diff, ytrain['TARGET'])
new_df=features_filtered_diff.drop(['timeseries__sample_entropy'], axis=1) # on supprime cette variable qui nous embête
```
