**Pré processing des données**

On a utilisé la libraire tsfresh pour faire ressortir les caractéristiques des séries temporelles. Je crée donc des nouveaux jeux de données avec les caractéristiques des séries temporelles sur lesquelles on peut appliquer des algorithmes de machine learning. 
Dans ce notebook, on crée :  
- le csv xtest_features.csv : caractéristiques extraites des séries temporelles du fichier xtest (échantillon test)
- le csv xtest_features_diff.csv : caractéristiques extraites des séries temporelles différenciées du fichier xtest (échantillon test)

```python
#pip install tsfresh
```

```python
from tsfresh import extract_features, select_features
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("xtest.csv")
xtest = data.set_index("ID")
xtest
```

**Création du dataframe formaté à la librairie tsfresh**

```python
t=[]
t.extend(np.arange(0,50))
time = t*11969
len(time)
```

```python
ID = []
for i in xtest.index:
    x=[i]*50
    ID.extend(x)
len(ID)
```

```python
Timeseries=[]
for row in xtest.itertuples() : 
    Timeseries.extend(row[2:])

len(Timeseries)
```

```python
df = pd.DataFrame(np.transpose([time,ID,Timeseries]),columns=['time','ID','timeseries'])
df
```

**Application de l'extraction des features (tsfresh)**

```python
extracted_features = extract_features(df, column_id="ID", column_sort="time")
```

```python
extracted_features
```

On récupère le nom des colonnes du fichier xtrain_features pour garder les mêmes colonnes dans le fichier test

```python
xtrain_features = pd.read_csv("xtrain_features.csv")
xtrain_features
```

```python
features_filtered = extracted_features[xtrain_features.columns[1:]]
features_filtered
```

```python
features_filtered.to_csv("xtest_features.csv") # création du fichier xtest
```

**Différenciation**

On extrait également les features mais la différence avec l'extraction précédente est que l'on extrait les features du jeu de données des séries temporelles différenciées.

```python
xtest_diff=xtest.iloc[:,1:].T.diff()[1:].T
xtest_diff
```

```python
xtest
```

**Création du dataframe formaté à la librairie tsfresh**

```python
t_diff=[]
t_diff.extend(np.arange(0,49))
time = t_diff*11969
len(time)
```

```python
ID_diff = []
for i in xtest_diff.index :
    x=[i]*49
    ID_diff.extend(x)
len(ID_diff)
```

```python
Timeseries_diff=[]
for row in xtest_diff.itertuples() : 
    Timeseries_diff.extend(row[1:])

len(Timeseries_diff)
```

```python
df_diff = pd.DataFrame(np.transpose([time,ID_diff,Timeseries_diff]),columns=['time','ID','timeseries'])
df_diff
```

```python
extracted_features = extract_features(df_diff, column_id="ID", column_sort="time")
```

```python
extracted_features
```

On récupère le nom des colonnes du fichier xtrain_features_diff pour garder les mêmes colonnes dans le fichier test

```python
xtrain_features = pd.read_csv("xtrain_features_diff.csv")
xtrain_features.columns[1:]
```

```python
features_filtered_diff = extracted_features[xtrain_features.columns[1:]]
features_filtered_diff
```

```python
features_filtered_diff.to_csv("xtest_features_diff.csv")
```
